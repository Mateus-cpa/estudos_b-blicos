### Apostas: Quando a preguiça e a ganância se encontram

Aquele que se apoderar de riquezas sem dar o valor da sua geração, tais quais tempo, energia, conhecimento, poderá não ter sabedoria para utilizá-las com sabedoria e poderá perder tudo da mesma forma.

> ⁴ Torna-se pobre aquele que lida com a mão negligente; mas a mão do diligente enriquece. (Provérbios  BKJ 10:4)

Muitos exemplos são pessoas que apostam de alguma forma, ou que confiam nessa sorte para enriquecer (loterias, cassinos, jogos de azar, jogos de apostas, e até mesmo a herança). Muito provavelmente, aqueles que não forem conscientizados do valor do trabalho e perderão tudo, voltando ao estado anterior.





## Vaidade

Admnistrar grande volume de bens não é para muitos. Quando você deposita tod seu amor no dinheiro, você deixa de ser o administrador e passa a cuidar deles como um pai. Se preocupa diariamente com os rendimentos, deixa de dedicar seu tempo em cuidar da santificação.

>¹⁰ E tudo quanto desejaram os meus olhos não lhes neguei, nem privei o meu coração de alegria alguma; mas o meu coração se alegrou por todo o meu trabalho, e esta foi a minha porção de todo o meu trabalho.
¹¹ E olhei eu para todas as obras que fizeram as minhas mãos, como também para o trabalho que eu, trabalhando, tinha feito, e eis que tudo era vaidade e aflição de espírito, e que proveito nenhum havia debaixo do sol. (Eclesiastes ACF 2:10,11)

Cada um que estiver nesta caminhada deverá avaliar no seu íntimo se a finalidade é apenas para mostrar para um grupo de pessoas, ou para se provar que teve sucesso. Será esse uma finalidade que te dará bom testemuho?

## Trabalho para Deus
Nossa obra como servos não se limita ao mundo material.

>¹ Eu sou a videira verdadeira, e meu Pai é o lavrador.
² Toda a vara em mim, que não dá fruto, a tira; e limpa toda aquela que dá fruto, para que dê mais fruto.
³ Vós já estais limpos, pela palavra que vos tenho falado.
⁴ Estai em mim, e eu em vós; como a vara de si mesma não pode dar fruto, se não estiver na videira, assim também vós, se não estiverdes em mim.
⁵ Eu sou a videira, vós as varas; quem está em mim, e eu nele, esse dá muito fruto; porque sem mim nada podeis fazer.
⁶ Se alguém não estiver em mim, será lançado fora, como a vara, e secará; e os colhem e lançam no fogo, e ardem.
⁷ Se vós estiverdes em mim, e as minhas palavras estiverem em vós, pedireis tudo o que quiserdes, e vos será feito.
⁸ Nisto é glorificado meu Pai, que deis muito fruto; e assim sereis meus discípulos.
⁹ Como o Pai me amou, também eu vos amei a vós; permanecei no meu amor.
¹⁰ Se guardardes os meus mandamentos, permanecereis no meu amor; do mesmo modo que eu tenho guardado os mandamentos de meu Pai, e permaneço no seu amor.
¹¹ Tenho-vos dito isto, para que o meu gozo permaneça em vós, e o vosso gozo seja completo. (João ACF 15:1-11)

Não podemos nos esquecer que somos um ramo de uma videira, participamos de uma obra maior. Então não deixe que seu trabalho o desvie dos planos de Deus, e nem que o que você faz seja uma atividade que não é lícita biblicamente falando.

> ⁹ Porque nós somos colaboradores de Deus; Vós sois lavoura de Deus e edifício de Deus. (1 Coríntios BKF 3:9)

Nós nunca podemos nos esquecer que fazemos parte de algo maior. Nós estamos Nele e Ele está em nós.

Fomos colhidos por Deus e estamos colaborando para a próxima colheita de homens.

## Excelência

Aqui somos provocados a sempre atender ao chamado, sendo como representantes de Deus nos nossos meios, mostrando que o trabalho deve ser realizado com zelo e indefectível.

²³ E tudo quanto fizerdes, fazei-o de todo o coração, como ao Senhor, e não aos homens,

Pois ao sermos exemplos, ao sermos referência no que fazemos, apontamos para Deus em nossas obras no mundo. Desta forma, mostramos o reflexo da obra de Deus.

¹² Sendo o vosso comportamento honesto entre os gentios, para que, apesar de falarem mal de vós, como falam de malfeitores, possam glorificar a Deus no dia da visitação, pelas boas obras que em vós hão de contemplar. (1 Pedro BKJ 2:12)

Queremos que as pessoas vejam Jesus através de nossas vidas e que possamos inspirar as pessoas ao nosso redor por que somos e por como agimos e pelo que fazemos.

Somos sal e luz nesta terra.

De nada adianta trabalhar sem cessar e dar aos perdulários, por falta de ensino como administrador. Não é um conhecimento inato do homem, quando não se tratar de um dom de sabedoria extraordinária.

A maior parte do conhecimento advém do exemplo e do estudo. Inspire os seus a terem o devido respeito e cuidado, para que não recaia nem em idolatria, nem em desprezo pela previdência.

# 4. Investimentos

## O primeiro investimento
1 Reis 3:5-13 (NVI)
O crescimento profissional e da renda gerada por ela é fruto de boas escolhas. Salomão escolheu a sabedoria antes de todas as riquezas.

## Reserva de emergência
José do Egito nos ensina sobre a regra dos 20%. Ele inspirou as dicas atuais da divisão 50-30-20, onde dividimos as despesas em 50% para necessidades, 30% para despesas não essenciais e 20% para reservas e investimentos.

Nunca se sabe o dia de amanhã, pois quanso o amanhã chega, precisamos estar preparados.

## Diversificação
Eclesiastes 11:2 (NVI):
> Reparta o que você tem com sete, até mesmo com oito, pois você não sabe que desgraça poderá cair sobre a terra.

Seus investimentos de longo prazo devem ser divididos em diferentes formas de ganhar dinheiro:
- Empresas (seja um empresa familiar, ações, fundos de investimentos)
- Imóveis (seja em imóveis físicos, cotas de fundos imobiliários, letras de crédito - LCA/LCI, certificados de recebíveis - CRA/CRI)
- Reservas de valor (dinheiro físico, outras moedas correntes, títulos governamentais, Certificados de Depósito Bancário - CDB)

Se colocar tudo em um único investimento aumentam os riscos. Pode gerar bons lucros, mas podem causar grandes prejuízos. Diversificar é se proteger.

Por fim, no Evangelho segundo Lucas, trago uma passagem para avaliar outra camada, além da intenção original de Jesus.

⁴⁷ Qualquer que vem a mim e ouve as minhas palavras, e as observa, eu vos mostrarei a quem é semelhante:
⁴⁸ É semelhante ao homem que edificou uma casa, e cavou, e abriu bem fundo, e pôs os alicerces sobre a rocha; e, vindo a enchente, bateu com ímpeto a corrente naquela casa, e não a pôde abalar, porque estava fundada sobre a rocha.
⁴⁹ Mas o que ouve e não pratica é semelhante ao homem que edificou uma casa sobre terra, sem alicerces, na qual bateu com ímpeto a corrente, e logo caiu; e foi grande a ruína daquela casa. (Lucas ACF 6:47-49)

Originalmente, Jesus se referia ao homem que tinha bases sólidas espirituais não se abalaria com intempéries.
O homem que aprende sobre o valor do trabalho e da aplicação dos conhecimentos adquiridos de forma consistente, poderá ver todo seu patrimônio ruir.

# 5. Propósito

Ter dinheiro não é o problema, mas sim o amor ao dinheiro:

1 Timóteo 6:10 (NVI)
>⁹ Os que querem ficar ricos caem em tentação, em armadilhas e em muitos desejos descontrolados e nocivos, que levam os homens a mergulharem na ruína e na destruição,
¹⁰ pois o amor ao dinheiro é raiz de todos os males. Algumas pessoas, por cobiçarem o dinheiro, desviaram-se da fé e se atormentaram a si mesmas com muitos sofrimentos.

Enriquecimento é fruto do amadurecimento dos passos anteriores: Planejamento, Receita, Despesa e Investimento.

Dinheiro é uma ferramenta, não a prioridade.

A pessoa que prioriza o dinheiro ela troca sua vida (tempo, saúde, relacionamento, paz e sono), e por que não dizer sua alma, por ele.
