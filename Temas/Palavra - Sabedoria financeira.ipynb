Um lenhador não recebe seu salário apenas por afiar seu machado. Apesar de este ser necessário para cortar a árvore e que um bom fio facilitará o trabalho duro, mas somente o planejamento não assegura a execução.

>⁴ Quem observa o vento, nunca semeará, e o que olha para as nuvens nunca segará.

## Ganância

Vamos tratar a ganância aqui como o excesso de ação para atingir o enriquecimento.

Aqui se aponta no novo testamento sobre as motivações que te levam ao desejo pelo enriquecimento. Como proposto nas fases antereios, quando você ele a Deus seus planos, você consulta tanto se isso é um desígnio de Deus quanto se a motivação que você tem é válida espiritualmente.

> ¹¹ A riqueza conseguida pela vaidade será diminuída, mas aquele que ajunta pelo trabalho crescerá. (Provérbios BKJ 13:11)

Se seu coração está no dinheiro, e não em fazer o bem com o que lhe foi confiado, você está sucumbindo à cobiça:

>¹⁰ Porque o amor ao dinheiro é a raiz de todos os males; e nessa cobiça alguns se desviaram da fé, e se traspassaram a si mesmos com muitas dores. (1 Timóteo ACF 6:10).

Pois quem ama o dinheiro, está o transformando em objeto de adoração. E adorar a qualquer criatura ou ídolos é quebrar o primeiro Mandamento, de Amar a Deus sobre todas as Coisas.

>²⁴ Nenhum homem pode servir a dois senhores; porque ou há de odiar um e amar o outro, ou se apegará a um e desprezará o outro. Não podeis servir a Deus e a Mamom. (Mateus BKJ 6:24)

Nessa passagem, Jesus aponta que devemos fazer escolhas sobre onde depositamos nosso coração e nossa mente, nosso tempo e nossa atenção. O dinheiro é um meio, não um fim. Se não houver um destino ou finalidade para seu dinheiro, a busca é incessante. Nunca será suficiente o que você ganha ou o que administra.

>¹⁰ Quem amar o dinheiro jamais dele se fartará; e quem amar a abundância nunca se fartará da renda; também isto é vaidade.

¹² Doce é o sono do trabalhador, quer coma pouco quer muito; mas a fartura do rico não o deixa dormir.

No Novo Testamento, Jesus aponta os perigos de enriquecimento sem direcionamento. Pois os tesouros de Deus não podem ser saqueados, não minam pecados no homem, mas sim seu engrandecimento.

>¹⁹ Não ajunteis tesouros na terra, onde a traça e a ferrugem tudo consomem, e onde os ladrões minam e roubam;

# 3. Despesa





## Gastos

A vida terrena é composta por despesas, viver em sociedade gera despesas que devem ser pagas corretamente, seja a outra pessoa, uma empresa ou o prórpio Estado. Morar, se alimentar, se deslocar, possuir bens. Ter gastos e dívidas são inerentes à vida humana.

Na passagem abaixo, mostra que até Jesus devia impostos. Veja bem, não é o mesmo que inadimplência, quando você deixa vencer a dívida.

²⁴ E, chegando eles a Cafarnaum, aproximaram-se de Pedro os que cobravam as dracmas, e disseram: O vosso mestre não paga as dracmas?
²⁵ Disse ele: Sim. E, entrando em casa, Jesus se lhe antecipou, dizendo: Que te parece, Simão? De quem cobram os reis da terra os tributos, ou o censo? Dos seus filhos, ou dos alheios?

²⁷ Mas, para que os não escandalizemos, vai ao mar, lança o anzol, tira o primeiro peixe que subir, e abrindo-lhe a boca, encontrarás um estáter; toma-o, e dá-o por mim e por ti. (Mateus ACF 17:24-27)

Quando provocado, Jesus disse: "Dai a César o que é de César e a Deus, o que é de Deus". Nessa frase dos Evangelho (Mateus 22:21), Ele ensina a distinguir obrigações civis (impostos, leis) da devoção espiritual.

E ainda Paulo disse aos Romanos:

> ⁷ Portanto, dai a cada um o que deveis: a quem tributo, tributo; a quem imposto, imposto; a quem temor, temor; a quem honra, honra.
⁸ A ninguém devais coisa alguma, a não ser o amor com que vos ameis uns aos outros; porque quem ama aos outros cumpriu a lei. (Romanos ACF 13:7-8)

Gerenciar nossas despesas é um dos pontos chaves para o enriquecimento. Viver abaixo do nosso nível de receita é fundamental:

>²⁰ Há tesouro desejável e azeite na habitação do sábio, mas o homem tolo gasta tudo. (Provérbios 21:20 BKJ)

Não é rico e pobre, mas o sábio e o tolo. Não é o tamanho do salário, é a adequação das depesas correntes (mensais) ao seu salário.

Se não sobra nada do seu salário após as despesas, significa que você está mantendo um nível de vida maior que deveria, ou que você deveria buscar novas formas de renda para suprir suas necessidades.


## Dívidas e endividamento

Vamos fixar o conceito de endividamento como um estágio mais complexo da dívida devida, pois neste caso a pessoa não tem condições de quitá-la. Quando a(s) pessoa(s) perde(m) que deveriam administrar o controle financeiro das suas despesas e entram no endividamento.

Deve-se aprender a diferenciar oportunidades de impulsos. O que você comprou lhe renderá melhores condições de prosperar? Ou apenas aumentará suas dívidas para satisfazer seu prazer?

Se envolver em um endividamento requer bastante estudo e cautela, pois podem se agravar com situações para as quais você não se preparou.

> ²⁶ Não estejas entre os que se comprometem, e entre os que ficam por fiadores de dívidas,
²⁷ Pois se não tens com que pagar, deixarias que te tirassem até a tua cama de debaixo de ti?
²⁸ Não removas os antigos limites que teus pais fizeram. (Provérbios ACF 22:26-28)

Digamos que uma pessoa quer um aparelho *smartphone* de última linha sem ter recursos para isso. Sem planejamento vai ao banco e pede um empréstimo, ou parcela no cartão.

No mês seguinte, ocorre um sinistro de saúde na família e você terá despesas inesperadas. Como resolver sem uma reserva para emergências, ou sem limite no cartão? Terá que recorrer a outro endividamento de pior qualidade, como o rotativo do cartão, o LIS (limite) da conta bancária, deixando-a no negativo ou ainda pedindo emprestado para uma financeira com juros altíssimos.

Veja que uma decisão errada desencadeou várias outras situações que se agravaram por conta de um ato impulsivo.

### Nome e honra

O cristão não deve se acomodar no endividamento. Quando se der conta de que não consegue pagar suas dívidas, primeiramente, deve se reconciliar com seu credor:

> ²³ Portanto, se trouxeres a tua oferta ao altar, e aí te lembrares de que teu irmão tem alguma coisa contra ti,
²⁴ Deixa ali diante do altar a tua oferta, e vai **reconciliar-te primeiro com teu irmão** e, depois, vem e apresenta a tua oferta.
²⁵ Concilia-te depressa com o teu adversário, enquanto estás no caminho com ele, para que não aconteça que o adversário te entregue ao juiz, e o juiz te entregue ao oficial, e te lancem na prisão.
²⁶ Em verdade te digo que de maneira nenhuma sairás dali enquanto não pagares o último centavo.  (Mateus ACF 5:23-26)

A sua palavra deve ter valor e ser referência, pois ela será a referência que as pessoas terão de um cristão.

>¹ Guarda o pé, quando entrares na Casa de Deus; chegar-se para ouvir é melhor do que oferecer sacrifícios de tolos, pois não sabem que fazem mal.
² Não te precipites com a tua boca, nem o teu coração se apresse a pronunciar palavra alguma diante de Deus; porque Deus está nos céus, e tu, na terra; portanto, sejam poucas as tuas palavras.
³ Porque dos muitos trabalhos vêm os sonhos, e do muito falar, palavras néscias.
⁴ Quando a Deus fizeres algum voto, não tardes em cumpri-lo; porque não se agrada de tolos. Cumpre o voto que fazes.
⁵ **Melhor é que não votes do que votes e não cumpras**. (Eclesiastes ARA 5:1-5)

Qual credibilidade você passa quando faz promessas sem a intenção de cumprir? Você deve ser o exemplo. Seu nome tem valor

>¹ Mais vale o bom nome do que as muitas riquezas;
  e o ser estimado é melhor do que a prata e o ouro. (Provérbios ARA 22:1)

Por fim, tenha a intenção real de amortizar e assim que possível quitar. Você pode estar prejudicando outra pessoa ou o relacionamento dela se ela não emprestou de forma clara e sem o interesse de retorno.

>¹ Melhor é a boa fama do que o unguento precioso, e o dia da morte, melhor do que o dia do nascimento. (Eclesiastes ARA 7:1)

As palavras de Paulo dá uma lição não só às viúvas, mas também a qualquer pessoa:

> ⁵ Ora, a que é verdadeiramente viúva e desamparada confia em Deus e continua em súplicas e orações noite e dia;
**⁶ mas a que vive em prazer, está morta enquanto ela vive.
⁷ Ordena estas coisas, para que elas sejam irrepreensíveis.**
⁸ Contudo, se alguém não tem cuidado dos seus e principalmente dos da sua casa, negou a fé e é pior do que um infiel. (1 Timóteo BKJ 5:5-8)

A pessoa que vive de seus desejos, não controla seus impulsos, vive uma vida de prazeres imediatos e não tem o cuidado para com os seus. Ela atende aos seus impulsos e se coloca em uma posição de vulnerabilidade, dependendo de empréstimos e se desonrando como um membro na sociedade.

Outras consequências serão tratadas abaixo.



### Sobre emprestar o nome (fiador) ou dinheiro

Salomão fala em Provérbios sobre ajudar pessoas sendo-lhe fiador:

> ¹ Filho meu, se ficaste por fiador do teu companheiro, se deste a tua mão ao estranho,
² E te deixaste enredar pelas próprias palavras; e te prendeste nas palavras da tua boca;
³ Faze pois isto agora, filho meu, e livra-te, já que caíste nas mãos do teu companheiro: vai, humilha-te, e importuna o teu companheiro.
⁴ Não dês sono aos teus olhos, nem deixes adormecer as tuas pálpebras.
⁵ Livra-te, como a gazela da mão do caçador, e como a ave da mão do passarinheiro. (Provérbios ACF 6:1-5)

Pensando nos dias de hoje, se uma pessoa precisa de dinheiro, o caminho mais correto é pedir a uma instituição que tem essa finalidade.

Quando comecei a trabalhar, li em um livro a seguinte frase: *antes de pedir dinheiro a um amigo, pense de qual dos dois você precisa mais*. Um relacionamento que envolve dinheiro pode ser arruinada com muita facilidade. O que pediu emprestado fica desconfiado, se afasta, sente-se vigiado pelos seus gastos. O relacionamento se verticaliza, ou seja, o fiador é colocado pelo outro como uma pessoa superior, a quem ele deve prestar contas.


As lições que eu aprendi são:

- Não se deve emprestar dinheiro nem a parente, imagine a estranhos.

- Nunca empreste um dinheiro que você precisa. Na pior das hipóteses, tenha uma reserva de empréstimos que você não tenha necessidade de receber de volta.

- Emprestar dinheiro é uma forma de dar corda para uma pessoa se enforcar. Se puder não emprestar, ao invés dê conselhos, pois o dinheiro só vai acomodar a outra pessoa. Com conselhos ela pode tentar formas criativas de ganhar um dinheiro extra.

- Se você emprestou dinheiro e não quer perder o relacionamento, perdoe a dívida. É um fardo a menos.

>²³ Por isso o reino dos céus pode comparar-se a um certo rei que quis fazer contas com os seus servos;
²⁴ E, começando a fazer contas, foi-lhe apresentado um que lhe devia dez mil talentos;
²⁵ E, não tendo ele com que pagar, o seu senhor mandou que ele, e sua mulher e seus filhos fossem vendidos, com tudo quanto tinha, para que a dívida se lhe pagasse.
²⁶ Então aquele servo, prostrando-se, o reverenciava, dizendo: Senhor, sê generoso para comigo, e tudo te pagarei.
²⁷ Então o senhor daquele servo, movido de íntima compaixão, soltou-o e perdoou-lhe a dívida.
²⁸ Saindo, porém, aquele servo, encontrou um dos seus conservos, que lhe devia cem dinheiros, e, lançando mão dele, sufocava-o, dizendo: Paga-me o que me deves.
²⁹ Então o seu conservo, prostrando-se a seus pés, rogava-lhe, dizendo: Sê generoso para comigo, e tudo te pagarei.
³⁰ Ele, porém, não quis, antes foi lançá-lo na prisão, até que pagasse a dívida.
³¹ Vendo, pois, os seus conservos o que acontecia, contristaram-se muito, e foram declarar ao seu senhor tudo o que se passara.
³² Então o seu senhor, chamando-o à sua presença, disse-lhe: Servo malvado, perdoei-te toda aquela dívida, porque me suplicaste.
³³ Não devias tu, igualmente, ter compaixão do teu conservo, como eu também tive misericórdia de ti?
³⁴ E, indignado, o seu senhor o entregou aos atormentadores, até que pagasse tudo o que lhe devia.
³⁵ Assim vos fará, também, meu Pai celestial, se do coração não perdoardes, cada um a seu irmão, as suas ofensas. (Mateus ACF 18:23-35)






## Prejuízos financeiros e espirituais da dívida

A Bíblia nos alerta em Provérbios:
> O rico domina sobre o pobre; quem toma emprestado é escravo de quem empresta. (Provérbios NVI 22:7)

Escravo bem do hebraico *eved*, substantivo do verbo *avad* que significa trabalhar ou servir. O mesmo termo utilizado sobre os escravos hebreus no Egito.

Permanecer endividado leva a pessoa a se tornar uma bola de neve e a sua vida vai mudando se você inconsientemente. Você passa a evitar pessoas e lugares. Pod eescandalizar, virar alvo de comentários. A reputação se corrói. E isso se torna uma consequência.

> ²⁴ As mãos diligentes governarão, mas os preguiçosos acabarão em trabalho forçado. (Provérbios NVI 12:24)

Seu salário passa a servir para pagar os juros da dívida, você perde qualidade de vida e paz, perde seu tempo para recompensar o diligente.

>²⁶ Não estejas entre os que se comprometem

## Credor

Se você já emprestou, melhor que tenha dado e que não conte com o dinheiro de volta. Quando uma pessoa pede dinheiro emprestado:

1. Foi uma emergência inesperada acima de suas reservas de emergência;
2. A pessoa precisa de recursos para realizar um projeto e pretende multiplicar o dinheiro;
3. Ela não se organizou financeiramente para emergências;
4. Ela vive num padrão acima da sua renda.
5. Não possui certidão positiva (leia-se nome limpo) para solicitar um empréstimo formalmente para uma instituição financeira.

Em qualquer hipótese, ao emprestar não espere que você receba de volta. Contabilize como uma bênção o retorno, pois se envolver nessa situação pode perturbar a sua paz.
