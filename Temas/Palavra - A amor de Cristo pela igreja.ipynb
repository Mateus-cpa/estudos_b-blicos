# A amor de Cristo pela igreja

Efésios 5:22-25: A mulher deve ser submissa ao homem, e o homem deve amar a mulher como Jesus amou a igreja.

Jesus:

* Se despiu de seus direitos e poderes pela igreja;  
* Chamou cada discípulo pelo seu nome, não só por quem era, mas por quem poderia ser;  
* Ensinou e delegou seus poderes aos discípulos, pois sabia que seu tempo entre eles era finito;  
* Entrou em Jerusalém montado em um jumento para cumprir a profecia, mesmo sabendo que iria ser sacrificado para espiar os pecados da humanidade;  
* Jesus lavou os pés dos discípulos para ensinar humildade pelo exemplo (Jo 13);  
* Lucas 22:44: Orou tão ardentemente no Getsêmani, que seu suor virou sangue, pois sabia tudo que aconteceria na sequência, mesmo assim não fugiu;   
* Prometeu que seria sacrificado e a partir dali a dívida do homem Adão teríamos comunhão com Deus;  
* Alertou aos discípulos que, após sua partida, encontrariam guerra, perseguição e que seriam odiados pelo amor a Jesus, mas prometeu que prepararia um lugar em sua morada e que os levaria para si.


Jesus preparou a igreja para o mundo após a sua partida, que se mantivesse firme até seu reencontro.
