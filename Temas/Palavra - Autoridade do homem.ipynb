# **Autoridade pela liderança**

Desde o tempo do Êxodo, Moisés sentiu-se sobrecarregado. Num primeiro momento, seu sogro Jetro lhe aconselhou para descentralizar a responsabilidade de julgar as pequenas causas aos anciãos. Num segundo momento, Deus instruiu na nomeação do Sumo-sacerdote, dos sacerdotes e dos levitas, além da instituição dos juízes.

Assim, se formaram os diferentes papeis na sociedade religiosa: o sacerdote, o juiz (posteriormente foi substituído pelo rei) e o profeta.

## *O sacerdote*
É aquele que intercede por aqueles que estão sob sua responsabilidade. Via-se muitos líderes sempre se colocando entre Deus e aqueles sobre sua responsabilidade.

Em Gênesis 18:19 Abraão **intercede** por seus parentes. Igualmente por várias vezes foi à presença de Deus para em favor do povo hebreu.

Também em outro ponto do Antigo Testamento:
>Jó chamava seus filhos e os santificava. (Jó 1:5)

Moisés, também intercedeu pelo povo no deserto nos livros do Pentateuco, quando esses pecavam e Deus quis eliminá-los.

Após a construção do tabernáculo, Deus instruiu Moisés a ungir Arão, seu irmão, como Sumo-Sacerdote. Tudo foi minunciosamente descrito, mas dentre os detalhes das veste, destaco o trecho abaixo:

>⁶ Prenderam as pedras de ônix em filigranas de ouro e nelas **gravaram os nomes dos filhos de Israel**, como um lapidador grava um selo.  
⁷ Então as **costuraram nas ombreiras** do colete sacerdotal, como pedras memoriais para os filhos de Israel, como o Senhor tinha ordenado a Moisés. (Êxodo 39:6-7)

O Sumo-Sacerdote carrega sobre os ombros o peso das 12 tribos de Israel, sendo o responsável por interceder por eles.

### *O profeta*

O profeta é aquele que transmite as palavras e os ensinamentos de Deus. Dentre os mais conhecidos, temos o próprio Moisés. Ele recebia com frequência as orientações de Deus, ele as transmitia ao povo hebreu, e também as escreveu nos primeiros livros da Bíblia para que ficassem registradas.

>Jó 4:3: **Ensinava** a muitos e fortalecia as mãos fracas.

Em vários pontos da bíblia, no período patriarcal, Os chefes de família exerciam suas funções de ensinamento.

Posteriormente, no peíodo de reinado quando os papel do profeta era impostantíssimo para orientar o rei no seu governo e muitas vezes traziam palavras duras de admoestação, como fizeram Natã:

>⁹ Por que, pois, desprezaste a palavra do Senhor, fazendo o que era mau perante ele? A Urias, o heteu, feriste à espada; e a sua mulher tomaste por mulher, depois de o matar com a espada dos filhos de Amom. (2 Samuel ARA 12:9)

Assim o fez também o profeta Elias:

>¹ E Elias, o tisbita, que era dos habitantes de Gileade, disse a Acabe: Como vive o SENHOR Deus de Israel, diante do qual me ponho de pé, não haverá orvalho, nem chuva nestes anos, senão segundo a minha palavra. (1 Reis BKJ 17:1)

O profeta desenvolve os ouvidos espirituais para transmitir a sabedoria de Deus aos seus.

## *o rei*

O rei é aquele que segue, viabiliza e coloca em prática a lei. Seria o equivalente ao Poder Executivo.

O Rei é o exemplo. É aquele que segue à frente.

Num primeiro momento, o papel executivo da sociedade hebraica eram os 70 anciões (Números 11:16-17), depois os juízes (Juízes 2:16) e depois um rei sobre Israel (1 Samuel 10:1)

> ¹⁵ Porém, se vos parece mal servir ao Senhor, escolhei, hoje, a quem sirvais: se aos deuses a quem serviram vossos pais que estavam dalém do Eufrates ou aos deuses dos amorreus em cuja terra habitais. **Eu e a minha casa serviremos ao Senhor**. (Josué ARA 24:15)

O rei é o líder espiritual em quem os demais acreditam, se inspiram e seguem. Ele toma as decisões baseado em sabedoria e conselhos e assume a responsabilidade pelas consequências de suas escolhas.



# Jesus e a Autoridade pela submissão

Jesus é o profeta, o sacerdote e o rei perfeito. No Novo Testamento, a estrutura de igreja foi descentralizado do Templo para o indivíduo.

>²¹ Disse-lhe Jesus: Mulher, podes crer-me que a hora vem, quando nem neste monte, nem em Jerusalém adorareis o Pai.
²² Vós adorais o que não conheceis; nós adoramos o que conhecemos, porque a salvação vem dos judeus.
²³ Mas vem a hora e já chegou, em que os **verdadeiros adoradores adorarão o Pai em espírito e em verdade**; porque são estes que o Pai procura para seus adoradores. (João ARA 4:21-23)

Em Mateus, o centurião reconheceu a autoridade de Cristo quando pediu que este curasse seu servo
>⁸ Mas o centurião respondeu: Senhor, não sou digno de que entres em minha casa; mas apenas manda com uma palavra, e o meu rapaz será curado.
⁹ Pois **também** eu sou homem sujeito à autoridade, tenho soldados às minhas ordens e digo a este: vai, e ele vai; e a outro: vem, e ele vem; e ao meu servo: faze isto, e ele o faz. (Mateus ARA 8:8,9)

Recebemos esse conhecimento valiosíssimo sobre submissão no curso de discipulado. A autoridade do centurião é consequência de uma cadeia de autoridade que vinha desde o imperador:

- Os nomeados pelo imperador  
- Os nomeados por estes como segundos líderes  
- E uma cadeia de autoridade até chegar aos centuriões nas milícias, que cuidam da manutenção da ordem nas terras conquistadas pelos romanos.

Sua autoridade vem da submissão, as pessoas acreditavam que ele dizia a própria palavra do imperador, como se fossem a mesma pessoa.

Da mesma forma Jesus **se coloca** e também **a nós** na sua linha de autoridade espiritual:

- João 10:30 "Eu e o Pai somos um"  
- João 14:28: Ouvistes que eu vos disse: vou e volto para junto de vós. Se me amásseis, alegrar-vos-íeis de que eu vá para o Pai, pois o Pai é maior do que eu.  
- I Cor 11:3: “Quero, porém, que entendam que o cabeça de todo homem é Cristo, o cabeça da esposa é o marido e o cabeça de Cristo é Deus".

O Senhor cobrará do homem o sucesso espiritual de sua casa.
