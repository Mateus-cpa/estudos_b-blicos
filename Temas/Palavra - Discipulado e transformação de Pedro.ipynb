# 1. Simão, o pescador

Jesus em seu primeiro contato com Simão e André, filhos de Jonas e os chamou:

>¹⁶ Andando à beira do mar da Galiléia, Jesus viu Simão e seu irmão André lançando redes ao mar, pois eram pescadores.
¹⁷ E disse Jesus: "Sigam-me, e **eu os farei** pescadores de homens".
Marcos NVI 1:16,17

*Outras referências: Mt 4:19, Lc 5:1-11, Jo 1:51.*

Eles não conheciam Jesus pessoalmente. Talvez André tenha ouvido falar sobre ele, pois era seguidor de João Batista. Mas Jesus conhece cada um profundamente, pois ele nos sonda (Salmos 139).

O pastor chama suas ovelhas pelo nome e elas o seguem (João 10:3-4).

## Impetuosidade de Pedro

Infere-se na Bíblia que Pedro era:
- fisicamente: mais velho que os demais apóstolos, tendo idade civil para ser cobrado o imposto de manutenção do templo (Mateus 17:24-27);
- civilmente: casado, pois Jesus curou sua sogra;
- psicologicamente: ousado, impetuoso e reativo, como mostrado em eventos futuros.

# 2. Chamado à transformação


## Deus concede poderes a Moisés
Dentro deste contexto, é possível observar analisar o aspecto simbólico dos poderes que Deus ensinou a Moisés para mostrar sinais de prova ao povo hebreu de que ele está agindo em nome de Deus (Êxodo 4:1-17).

Neste ponto da história, Moisés foi criado por egípcios e, depois de matar um deles, fugiu para Midiã, onde tinha passado 40 anos como pastor.

**Bordão em serpente 🐍**: Moisés **fugia** da serpente, porque ela simboliza o Egito, era como encarar seu passado lá.

Na segunda instrução de recolhê-la, e tomá-la pela cauda é um ato de confiança extrema em Deus, pois não se pega uma cobra pela cauda, visto que seria picado. Esse ato é como reconhecer que Deus pode nos auxiliar a lidar com nosso passado.

**Mão leprosa 🤚**: Ao colocar a mão no peito e retirá-la, Moisés encarou a corrupção que havia no coração dos homens com a mão leprosa, e a sua própria ao convivendo com a cultura egípcia.

Mas no segundo comando, tornou a colocá-la e depois foi curada. Mais uma vez Deus mostra que é capaz de nos curar a feridas

Foi chamado à sua transformação pessoal, para então transformar seu povo.

Você o escuta, mas não pode dizer de onde vem nem para onde vai. Assim acontece com todos os nascidos do Espírito”.

## A morte do Eu

Somos seres humanos com história, manias, tiques e fobias. Todos carregamos um fardo do passado em alguma medida, mas para mais profunda transformação, é necessário a morte verdadeira do Eu antigo.

A morte do Eu é o primeiro passo para o discipulado. E ocorre em 3 passos:
1. Reconhecimento da sabedoria de Deus sobre nós. (salmos 139)
2. Negar-se a ou Esvaziar-se de si mesmos

Sobre isso, disse Paulo aos Filipenses:

>⁵ Seja a atitude de vocês a mesma de Cristo Jesus,
⁶ que, embora sendo Deus, não considerou que o ser igual a Deus era algo a que devia apegar-se;
⁷ mas **esvaziou-se a si mesmo, vindo a ser servo**, tornando-se semelhante aos homens.
⁸ E, sendo encontrado em forma humana, humilhou-se a si mesmo e foi obediente até à morte, e morte de cruz! (Filipenses 2:5-8)

Também Jesus disse aos que querem o seguir:

> 23 Jesus dizia a todos: "Se alguém quiser acompanhar-me, **negue-se a si** mesmo, tome diariamente a sua cruz e siga-me. (Lucas NVI 9:23)

E ainda Deus narrou a Moisés:

> Comereis a colheita velha para dar lugar ao novo, tiraremos fora o velho. (Levítico ARA 26:10)

3. Encher-se de água viva
Após esvaziar-se é possível encher-se de novo conteúdo. Isso é um **sinal profundo** de transformação. Jesus convida a mulher samaritana, bem como em outra ocasião:

>³⁷ No último e mais importante dia da festa, Jesus levantou-se e disse em alta voz: "Se alguém tem sede, venha a mim e beba. ³⁸ Quem crer em mim, como diz a Escritura, do seu interior fluirão rios de água viva". (João NVI 7:37-38)

E assim, nos tornaremos nova criatura em Jesus e nunca mais teremos sede.
Ainda na carta de Paulo aos Gálatas:

>Já estou crucificado com Cristo; e vivo, não mais eu, mas Cristo vive em mim (Gálatas ACF 2:20)






## A transformação mineral de Pedro

O nome carrega um significado sobre as circunstâncias do nascimento do bebê, seja sobre a situação do parto, dificuldades da gravidez, situação política, condição do pai, ou até homenagem à família ou a um profeta. Ele pode influenciar inclusive em nossa personalidade.

Em vários momentos na Bíblia, Deus deu novo nome indicando sua transformação de vida, tais quais Abraão, Sara, Israel.

Quando Jesus e seus discípulos estavam em Cesareia de Filipe, provocou seus discípulos sobre quem era o Filho do Homem:

> ¹⁵ "E vocês? ", perguntou ele. "Quem vocês dizem que eu sou? "
¹⁶ Simão Pedro respondeu: "Tu és o Cristo, o Filho do Deus vivo".
¹⁷ Respondeu Jesus: "Feliz é você, Simão, filho de Jonas! Porque isto não lhe foi revelado por carne ou sangue, mas por meu Pai que está nos céus.
¹⁸ E eu lhe digo que você é Pedro, e sobre esta pedra edificarei a minha igreja, e as portas do Hades não poderão vencê-la.
¹⁹ Eu lhe darei as chaves do Reino dos céus; o que você ligar na terra terá sido ligado nos céus, e o que você desligar na terra terá sido desligado nos céus".
Mateus NVI 16:15-19

Este foi um ponto crucial na história de Pedro, pois neste ponto ele recebeu uma missão de maior responsabilidade, exigindo uma mudança interna como um todo.

Mas Simão Pedro, tinha altos e baixos, mas era um homem em constante aprimoramento. Se fôssemos colocar as atitudes de Pedro num gráfico, ficaria algo assim:

Gráfico Pedro (melhorado).jpg

Nossa transformação é algo contínuo e possui altos e baixos, mas o principal é que criemos maturidade para nos reerguermos, nos perdoarmos como pessoas falhas, e retomar nossa cruz. **O Simão que andou sobre as águas era mais maduro na fé do que o pescador que foi ajudado por Jesus.**

Esse gráfico mostra que Jesus **não escolheu os homens perfeitos** para serem seus discípulos, mesmo Simão Pedro, que recebeu a maior responsabilidade. Mas Deus nos conhece no íntimo da alma e conhece nossas dificuldades.

# 3. A misericórdia do Senhor

Pedro viveu 3 anos ao lado de Jesus e presenciou toda sorte de milagres. Mesmo assim, na Última Ceia:

>³³ Mas Pedro, respondendo, disse-lhe: Ainda que todos se escandalizem em ti, eu nunca me escandalizarei.
³⁴ Disse-lhe Jesus: Em verdade te digo que, nesta mesma noite, antes que o galo cante, três vezes me negarás.
³⁵ Disse-lhe Pedro: Ainda que me seja mister morrer contigo, não te negarei. E todos os discípulos disseram o mesmo.
(Mateus NVI 26:33-35)

Pedro olhou nos olhos de Jesus e afirmou que não o negaria, mas o fez.

Ele sabe o coração das pessoas e que elas são fracas e que vão pecar, mas a **Morte do Eu** é um processo contínuo. Quando ressuscitou, em uma das oportunidades que estiveram juntos, deu os 3 perdões a Pedro:

>¹⁵ Depois de comerem, Jesus perguntou a Simão Pedro: "Simão, filho de João, você me ama realmente mais do que estes? " Disse ele: "Sim, Senhor, tu sabes que te amo". Disse Jesus: "Cuide dos meus cordeiros".
¹⁶ Novamente Jesus disse: "Simão, filho de João, você realmente me ama? " Ele respondeu: "Sim, Senhor tu sabes que te amo". Disse Jesus: "Pastoreie as minhas ovelhas".
¹⁷ Pela terceira vez, ele lhe disse: "Simão, filho de João, você me ama? " Pedro ficou magoado por Jesus lhe ter perguntado pela terceira vez "Você me ama? " e lhe disse: "Senhor, tu sabes todas as coisas e sabes que te amo". Disse-lhe Jesus: "Cuide das minhas ovelhas. ¹⁸ Digo-lhe a verdade: Quando você era mais jovem, vestia-se e ia para onde queria; mas quando for velho, estenderá as mãos e outra pessoa o vestirá e o levará para onde você não deseja ir".
**João NVI 21:15-18**
